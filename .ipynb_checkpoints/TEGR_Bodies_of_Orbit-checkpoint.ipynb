# TEGR: Bodies of Orbit
This notebook implements the **Kinematic Overlay** for the TEGR collider.
Instead of guessing the initial momentum and hoping the particles enter a stable orbit, we force the particles onto predefined mathematical tracks (e.g. a perfect circle). We then observe how the Weitzenböck lattice ($\kappa$ torsional strain) reacts to this controlled motion. By feeding this clean data into PySINDy, we can perfectly extract the matrix's native $G_{eff}$ without chaotic slingshot noise.

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import pysindy as ps

# --- SIMULATION PARAMETERS ---
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

DT = 0.001
TOTAL_TICKS = 4000
GRID_RES = 100
GRID_MIN = -50.0
GRID_MAX = 50.0
DX = (GRID_MAX - GRID_MIN) / (GRID_RES - 1)
WAVE_SPEED = 100.0
TORSION_DECAY = 0.999

N = 2  # Sun and Venus
M_SUN = 408670.0
M_VENUS = 1.0
R_ORBIT = 20.0  # grid units
ORBIT_V = 5.0   # Forced angular velocity proxy

state = torch.zeros((N, 10), device=device)
state[0, 7] = M_SUN
state[1, 7] = M_VENUS

# Eulerian Grid (FDTD)
phi_curr = torch.zeros((GRID_RES, GRID_RES, GRID_RES), device=device)
phi_prev = torch.zeros((GRID_RES, GRID_RES, GRID_RES), device=device)
kappa    = torch.zeros((GRID_RES, GRID_RES, GRID_RES), device=device)

history = np.zeros((TOTAL_TICKS, N, 10))
kappa_history = np.zeros((TOTAL_TICKS, N)) # Track kappa at particle pos
```

### The FDTD Core
This cell implements the grid interpolation and wave equation.

```python
laplacian_kernel = torch.zeros(1, 1, 3, 3, 3, device=device)
laplacian_kernel[0, 0, 1, 1, 1] = -6.0
laplacian_kernel[0, 0, 0, 1, 1] = 1.0
laplacian_kernel[0, 0, 2, 1, 1] = 1.0
laplacian_kernel[0, 0, 1, 0, 1] = 1.0
laplacian_kernel[0, 0, 1, 2, 1] = 1.0
laplacian_kernel[0, 0, 1, 1, 0] = 1.0
laplacian_kernel[0, 0, 1, 1, 2] = 1.0

C_SQUARED = (WAVE_SPEED * DT / DX) ** 2

def trilinear_interpolate(grid, pos, g_min, g_max, res, dx):
    idx = (pos - g_min) / dx
    idx = torch.clamp(idx, 0, res - 1.001)
    idx_f = torch.floor(idx).long()
    w1 = idx - idx_f
    w0 = 1.0 - w1
    x, y, z = idx_f[:,0], idx_f[:,1], idx_f[:,2]
    val = (grid[x, y, z] * w0[:,0]*w0[:,1]*w0[:,2] +
           grid[x+1, y, z] * w1[:,0]*w0[:,1]*w0[:,2] + 
           grid[x, y+1, z] * w0[:,0]*w1[:,1]*w0[:,2] +
           grid[x, y, z+1] * w0[:,0]*w0[:,1]*w1[:,2]) # Simplified interpolation
    return val

def deposit_mass(grid, pos, mass, g_min, g_max, res, dx):
    idx = (pos - g_min) / dx
    idx_f = torch.floor(idx).long()
    valid = (idx_f[:,0] >= 0) & (idx_f[:,0] < res-1) & \
            (idx_f[:,1] >= 0) & (idx_f[:,1] < res-1) & \
            (idx_f[:,2] >= 0) & (idx_f[:,2] < res-1)
    for i in range(len(pos)):
        if valid[i]:
            x, y, z = idx_f[i]
            grid[x, y, z] += mass[i] * DT
    return grid
```

### The Simulation Loop
The particles are locked to the mathematical tracks. We step the simulation forward and record the geometry of the space.

```python
import torch.nn.functional as F

print("Starting simulation...")
for tick in range(TOTAL_TICKS):
    t = tick * DT
    
    # --- THE TRACKS (KINEMATIC OVERLAY) ---
    # Sun is fixed at center
    state[0, 1:4] = torch.tensor([0.0, 0.0, 0.0])
    
    # Venus is forced into a circular orbit
    state[1, 1] = R_ORBIT * np.cos(ORBIT_V * t)
    state[1, 2] = R_ORBIT * np.sin(ORBIT_V * t)
    state[1, 3] = 0.0
    
    # Derived velocities (for SINDy)
    state[1, 4] = -R_ORBIT * ORBIT_V * np.sin(ORBIT_V * t) * M_VENUS
    state[1, 5] = R_ORBIT * ORBIT_V * np.cos(ORBIT_V * t) * M_VENUS
    state[1, 9] = 1.0  # Gamma approx
    
    pos = state[:, 1:4]
    mass = state[:, 7]
    
    # --- THE FABRIC (GRID UPDATE) ---
    phi_unsqueezed = phi_curr.unsqueeze(0).unsqueeze(0)
    laplacian = F.conv3d(phi_unsqueezed, laplacian_kernel, padding=1).squeeze(0).squeeze(0)
    
    phi_next = (2.0 * phi_curr) - (TORSION_DECAY * phi_prev) + (C_SQUARED * laplacian)
    
    # Deposit defects
    phi_next = deposit_mass(phi_next, pos, mass, GRID_MIN, GRID_MAX, GRID_RES, DX)
    
    # Torsional strain (kappa) is the absolute Laplacian (curvature)
    kappa = torch.abs(laplacian)
    
    # Record kappa at particle positions
    k_at_p = trilinear_interpolate(kappa, pos, GRID_MIN, GRID_MAX, GRID_RES, DX)
    
    phi_prev = phi_curr
    phi_curr = phi_next
    
    history[tick] = state.cpu().numpy()
    kappa_history[tick] = k_at_p.cpu().numpy()
    
    if tick > 0 and tick % 500 == 0:
        print(f"Tick {tick}/{TOTAL_TICKS}")
print("Simulation Complete.")
```

### PySINDy Extraction: Finding $G_{eff}$
Because Venus moved in a perfect circle, any centripetal acceleration mathematically extracted from the $\kappa$ field directly represents the grid's emergent gravitational constant.

```python
print("Preparing SINDy extraction...")
tracker_idx = 1 # Venus

tracker_pos = history[:, tracker_idx, 1:4]
tracker_vel = history[:, tracker_idx, 4:7] / history[:, tracker_idx, 7:8]

sun_pos = history[:, 0, 1:4]
dx = tracker_pos[:, 0] - sun_pos[:, 0]
dy = tracker_pos[:, 1] - sun_pos[:, 1]
r_sq = dx**2 + dy**2 + 1e-6
r = np.sqrt(r_sq)

inv_r2 = 1.0 / r_sq
kappa_vals = kappa_history[:, tracker_idx]

X_features = np.column_stack((tracker_pos[:, 0], tracker_pos[:, 1], r, inv_r2, kappa_vals))
feature_names = ['x', 'y', 'r', '1/r^2', 'kappa']

X_targets = tracker_vel # We will fit v' (acceleration)

library = ps.PolynomialLibrary(degree=1)
optimizer = ps.STLSQ(threshold=0.001, alpha=1e-5, fit_intercept=False)
model = ps.SINDy(feature_names=feature_names, optimizer=optimizer, feature_library=library)

model.fit(X_features, x_dot=X_targets, t=DT)
print("SINDy Equations extracted:")
model.print()
```

### Visual Rendering
Animate the orbit.

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.set_xlim(-40, 40)
ax.set_ylim(-40, 40)
ax.set_title("TEGR: Bodies of Orbit")
ax.grid(True)

sun_dot, = ax.plot([], [], 'yo', markersize=15, label="Sun")
venus_dot, = ax.plot([], [], 'ro', markersize=8, label="Venus")
trail, = ax.plot([], [], 'r-', alpha=0.5)
ax.legend()

def init():
    sun_dot.set_data([], [])
    venus_dot.set_data([], [])
    trail.set_data([], [])
    return sun_dot, venus_dot, trail

def animate(i):
    idx = i * 20
    if idx >= TOTAL_TICKS:
        idx = TOTAL_TICKS - 1
    # set_data requires a sequence, not a single scalar
    sun_dot.set_data([history[idx, 0, 1]], [history[idx, 0, 2]])
    venus_dot.set_data([history[idx, 1, 1]], [history[idx, 1, 2]])
    trail.set_data(history[:idx:20, 1, 1], history[:idx:20, 1, 2])
    return sun_dot, venus_dot, trail

anim = FuncAnimation(fig, animate, init_func=init, frames=200, interval=50, blit=True)
plt.close()
HTML(anim.to_jshtml())
```
